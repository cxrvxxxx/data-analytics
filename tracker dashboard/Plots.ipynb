## Cleaning Date/TIme

```python
berou_data = pd.read_excel('./Berou_Tracker.xlsx')
datetime_col = berou_data["Date/Time"]
datetime_col = [row for row in datetime_col]

date_col = []
time_col = []

for row in datetime_col:
    date = row.split(' ')
    date_col.append(date[0])
    time_col.append(date[1])
```

## Processing Date Column

```python
for i in range(len(date_col)):
    try:
        date_object = datetime.strptime(date_col[i], "%m/%d/%Y")
        date_col[i] = date_object.strftime("%Y-%m-%d")
    except ValueError:
        date_object = datetime.strptime(date_col[i], "%m//%d/%Y")
        date_col[i] = date_object.strftime("%Y-%m-%d")
```

## Processing Time Column

```python
for i in range(len(time_col)):
    if not (time_col[i].find(';') == -1):
        time_col[i] = time_col[i].replace(';', ':')

    if time_col[i].endswith('am'):
        time_col[i] = time_col[i][:-2]
        time_col[i] += ' AM'
    elif time_col[i].endswith('pm'):
        time_col[i] = time_col[i][:-2]
        time_col[i] += ' PM'
    else:
        time_col[i] += ' AM'
```

```python
for time in time_col:
    print(time)
```

```python
for i in range(len(time_col)):
    try:
        time_object = datetime.strptime(time_col[i], "%I:%M %p")
        time_col[i] = time_object.strftime("%H:%M")
    except:
        continue
```

```python
for time in time_col:
    print(time)
```

```python
duration_col = berou_data['Duration']

new_duration = []

for duration in duration_col:
    if duration.endswith('hrs'):
        value = duration[:-3]
        value = int(value)
        value = value * 60
    elif duration.endswith('mins'):
        value = duration[:-4]

    new_duration.append(value)
```
