## Loading the data

```python
df = pd.read_excel('./EmpasisBerouTabo_Tracker.xlsx')
df.head()
```

## Cleaning Date

```python
date_col = df["Date"].tolist()

for i in range(len(date_col)):
    if isinstance(date_col[i], str):
        date_col[i] = date_col[i].strip()

        if date_col[i] == '10//22/2023':
            date_col[i] = '10/22/2023'
        
        try:
            date_obj = datetime.strptime(date_col[i], "%m/%d/%Y")
        except ValueError:
            date_obj = datetime.strptime(date_col[i], "%m/%d/%Y %H:%M")
        finally:
            date_col[i] = date_obj.date()
    else:
        date_col[i] = date_col[i].date()

df["Date"] = date_col
df.head()
```

## Cleaning Time Started

```python
ts_col = df["Time Started"]

for i in range(len(ts_col)):
    if isinstance(ts_col[i], str):
        ts_col[i] = ts_col[i].strip()

        try:
            time_obj = datetime.strptime(ts_col[i], "%H:%M")
        except ValueError:
            time_obj = datetime.strptime(ts_col[i], "%H:%M:%S %p")

        ts_col[i] = time_obj.time()

    try:
        ts_col[i] = ts_col[i].time()
    except:
        pass

df["Time Started"] = ts_col
df.head()
```

## Cleaning Duration

```python
duration_col = df["Duration (mins)"]

for i in range(len(duration_col)):
    if isinstance(duration_col[i], str):
        if not duration_col[i].find(' and ') == -1:
            hour, min = duration_col[i].split(' and ')
            hour = int(hour[:-2])
            min = int(min.split(' ')[0])

            duration_col[i] = hour * 60 + min
            continue

        if duration_col[i].endswith('hr'):
            duration_col[i] = int(duration_col[i][:-2]) * 60
            continue
        
        if duration_col[i].endswith('hrs'):
            duration_col[i] = int(duration_col[i][:-3]) * 60
            continue
            
        if duration_col[i].endswith('mins'):
            duration_col[i] = int(duration_col[i][:-4])
            continue

        if duration_col[i].endswith('hour'):
            duration_col[i] = int(duration_col[i][:-4])
            continue

        if duration_col[i].endswith('hours'):
            duration_col[i] = int(duration_col[i][:-5])
            continue

df["Duration (mins)"] = duration_col
df.head()
```

## Cleaning Time Finished

```python
for i in range(len(df)):
    if not isinstance(df.iloc[i]["Duration (mins)"], int):
        df.iloc[i]["Duration (mins)"] = int(df.iloc[i]["Duration (mins)"])
        
    time_started = datetime.combine(datetime.today(), df.iloc[i]["Time Started"])
    df.iloc[i]["Time Finisheed"] = time_started + timedelta(minutes=df.iloc[i]["Duration (mins)"])

df.head()
```

```python
desc_col = df["Description"]

for i in range(len(desc_col)):
    desc_col[i] = desc_col[i].lower()

for i in range(len(desc_col)):
    if not desc_col[i].find('class') == -1:
        desc_col[i] = 'attended class'
        continue

    if not desc_col[i].find('social media') == -1:
        desc_col[i] = 'social media'
        continue

    if not desc_col[i].find('breakfast') == -1:
        desc_col[i] = 'had breakfast'
        continue

    if not desc_col[i].find('lunch') == -1:
        desc_col[i] = 'had lunch'
        continue

    if not desc_col[i].find('dinner') == -1:
        desc_col[i] = 'had dinner'
        continue

    if not desc_col[i].find('waking') == -1:
        desc_col[i] = 'woke up'
        continue

    if not desc_col[i].find('sleep') == -1:
        desc_col[i] = 'slept'
        continue

    if not desc_col[i].find('study') == -1:
        desc_col[i] = 'studied'
        continue

    if not desc_col[i].find('eat') == -1:
        desc_col[i] = 'ate'
        continue

    if not desc_col[i].find('driving') == -1:
        desc_col[i] = 'travelling'
        continue

    if not desc_col[i].find('riding') == -1:
        desc_col[i] = 'travelling'
        continue

    if not desc_col[i].find('going') == -1:
        desc_col[i] = 'travelling'
        continue

    if not desc_col[i].find('went') == -1:
        desc_col[i] = 'travelling'
        continue

    if not desc_col[i].find('relax') == -1:
        desc_col[i] = 'rested'
        continue

    if not desc_col[i].find('rest') == -1:
        desc_col[i] = 'rested'
        continue

    if not desc_col[i].find('games') == -1:
        desc_col[i] = 'played games'
        continue

    if not desc_col[i].find('cook') == -1:
        desc_col[i] = 'cooked'
        continue

    if not desc_col[i].find('school works') == -1:
        desc_col[i] = 'did school works'
        continue
```
