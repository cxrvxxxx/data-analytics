```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('umbrellas.csv')
df.head()
```

```python
plt.figure(figsize=(10,10))

x=df["Rainfall (mm)"]
y=df["Umbrellas sold"]

sns.regplot(
    data=df,
    x=x,
    y=y
)

plt.xlabel("Rainfall (mm))")
plt.ylabel("Umbrellas sold")

plt.show()
```
