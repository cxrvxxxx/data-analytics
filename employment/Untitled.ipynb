```python
import pandas as pd
import numpy as np
import seaborn as sns

from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("gdp_and_unemp.csv", index_col=0)
df.head()
```

```python
x = df["GDP"]
y = df["Unemp"]
```

```python
plt.figure(figsize=(6,4))

sns.regplot(
    x=x,
    y=y,
    line_kws={ "color": "#ef476f" },
    scatter_kws={"color": "#118ab2"}
)

plt.xlabel("GDP")
plt.ylabel("Unemployment")

plt.show()
```
