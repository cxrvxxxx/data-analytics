```python
reg1_data = pd.read_csv("regression1.csv")

x = reg1_data["height"]
y = reg1_data["weight"]
```

```python
plt.figure(figsize=(10, 6))

sns.regplot(
    x=x,
    y=y
)

plt.show()
```

```python
slope, intercept = np.polyfit(x, y, 1)
print(f"Equation: y = {slope}x + {intercept}")
```
