## Happiness Plot

```python
# Load the Data Frame
df = pd.read_csv('WHR2023.csv')

# Selecting the data to be displayed
x = df['Ladder score']
y = df['Freedom to make life choices']

# Creating a figure object
fig = plt.figure(figsize=[10,7])

# Creating the pie chart
plt.plot(x, y)

# Setting a title
plt.title('Happiness Plot')

# Show the chart
plt.show()
```

## Happiness Score vs Generosity and Perceptions of Corruption

```python
x = df['Ladder score'] # Create an array of x-values
y1 = df['Generosity']             # First set of y-values (sine function)
y2 = df['Perceptions of corruption']               # Second set of y-values (cosine function)

# Create the first line plot
plt.plot(x, y1, label='Generosity', linestyle='-', color='blue')

# Create the second line plot
plt.plot(x, y2, label='Perceptions of corruption', linestyle='-', color='green')

# Adding labels, title, and legend
plt.xlabel('Happiness Score')
plt.ylabel('')
plt.title('Happiness Score vs Generosity and Perceptions of Corruption')
plt.legend()

# Display the plot
plt.show()
```
