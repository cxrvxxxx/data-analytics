```python
import pandas as pd
import numpy as np

def main() -> None:
    df = pd.read_csv('ex1.csv')
    print(df)
    df['c'] = df['c'].fillna(np.mean(df['c']))
    print(df)

if __name__ == '__main__':
    main()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('ex2.csv')

x = df['x']
y = df['y']

plt.plot(x, y)

plt.xlabel('x - axis')
plt.ylabel('y - axis')

plt.title('My first graph!')

plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv('ex3.csv')

cars = df['cars']
data = df['data']

fig = plt.figure(figsize=[10,7])

plt.pie(data, labels=cars)
plt.title('Pie Chart')

plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv('ex3.csv')

cars = df['cars']
data = df['data']

fig = plt.figure(figsize=[10,7])

plt.bar(cars, data)
plt.title('Bar Graph')

plt.show()
```
