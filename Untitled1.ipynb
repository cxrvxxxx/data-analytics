```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import math
import statistics
print('Setup complete')
```

```python
# Reading the Dataset from fie
df = pd.read_excel('activity_tracker.xlsx')
df.sample(10)
```

```python
acts = df["Description"].unique()
times = [round(df[df["Description"] == act]["Duration (mins)"].mean(), 2) for act in acts]

graph_data = pd.DataFrame(times, acts)
graph_data = graph_data.rename(columns={0: "Duration"})
graph_data = graph_data.sort_values(by="Duration", ascending=False)

plt.figure(figsize=(10, 6))

ax = sns.barplot(
    x = graph_data["Duration"],
    y = graph_data.index
)

plt.ylabel("Activity")
plt.xlabel("Average Duration (mins)")
plt.title("Duration of Activities")

plt.show()
```
